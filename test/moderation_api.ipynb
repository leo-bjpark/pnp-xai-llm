```python
import os
from openai import OpenAI

key = "[REDACTED]"
client = OpenAI(api_key=key)

def get_moderation_predictions(texts):
    # ======================================== 
    # Call the moderation API
    MODEL = "omni-moderation-latest"
    response = client.moderations.create(
                model=MODEL,
                input=texts,
                     )
    # ======================================== 
    # Parse the response
    response_dict = response.to_dict()
    results = [] 
    for index, result in enumerate(response_dict['results']):
        predictions = response_dict['results'][0]['categories']
        scores = response_dict['results'][0]['category_scores']
        result = {
                # name: (prediction: bool, score: float)
        }
        for name in predictions:
            prediction = predictions[name]
            score = scores[name]
            assert prediction in [True, False]
            assert 0<=score <= 1
            result[name] = (prediction, score)
        results.append(result)
    return results
    
texts = ["How can I make a bomb?", 
        "One way to make a bomb is to use a chemical reaction to create a fireball."]
results = get_moderation_predictions(texts)
results
```
